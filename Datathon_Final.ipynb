RELIK Model

```python
# Other imports
import pandas as pd
from google.colab import files

# Entity Extraction Model
relik_entity1 = Relik.from_pretrained(
    "sapienzanlp/relik-entity-linking-small",
    device="cuda" if torch.cuda.is_available() else "cpu",
    precision="fp16" if torch.cuda.is_available() else "fp32",
    skip_metadata=True,
)

# Relationship Extraction Model
relik_relationship1 = Relik.from_pretrained(
    "relik-ie/relik-relation-extraction-large",
    device="cuda" if torch.cuda.is_available() else "cpu",
    precision="fp16" if torch.cuda.is_available() else "fp32",
    skip_metadata=True,
)
```

```python
# Text Extraction
uploaded = files.upload()
filename1 = list(uploaded.keys())[0]
data1 = pd.read_excel(filename1)
data1.rename(columns={data1.columns[0]: "Source"}, inplace=True)

texts1 = data1.iloc[:,1]

# store results
extracted_entities1 = []
extracted_relationships1 = []

for text in texts1:
  relik_entity_results1: RelikOutput = relik_entity1(text)
  extracted_entities1.append(relik_entity_results1.spans)

  relik_relationship_results1 = RelikOutput = relik_relationship1(text)
  extracted_relationships1.append(relik_relationship_results1.triplets)

output_df = data1.copy()
output_df['entities'] = [str(entity) for entity in extracted_entities1]
output_df['relationships'] = [str(relationship) for relationship in extracted_relationships1]

# Save to a new CSV file
output_file = "relikALL.csv"
output_df.to_csv(output_file, index=False)

# Download the CSV file
files.download(output_file)
```

Processing Relik Data

```python
import re
import pandas as pd
from google.colab import files
from rapidfuzz import fuzz
import pycountry  # For country standardization

# Step 1: Upload and Load Relik Data Files
print("Upload Relik Data Files (multiple files allowed):")
uploaded_files = files.upload()  # Allow multiple files

# Initialize a list to hold all DataFrames
dataframes = []

# Loop through uploaded files and read each into a DataFrame
for file_name in uploaded_files.keys():
    df = pd.read_csv(file_name)
    df['CSV'] = file_name  # Add a column to track which file the data came from
    dataframes.append(df)

# Combine all uploaded DataFrames into one
relik_data = pd.concat(dataframes, ignore_index=True)

# Dynamically identify the first column name (e.g., "PDF Path" or "Source")
source_column_name = relik_data.columns[0]

# Step 2: Upload and Load Demonyms File
print("Upload Demonyms File (demonyms.csv):")
demonyms_file = files.upload()  # Upload demonyms.csv

# Load demonyms.csv, explicitly setting column names
demonyms_filename = list(demonyms_file.keys())[0]  # Get uploaded file name
demonyms_df = pd.read_csv(demonyms_filename, names=["Demonym", "Country"], header=None)

# Convert to dictionary for fast lookups
demonyms_mapping = dict(zip(demonyms_df["Demonym"].str.lower(), demonyms_df["Country"]))

# Step 3: Parse Relik Relationships
def parse_relik_relationships(entry):
    pattern = r"Triplets\(subject=Span.*?text='(.*?)'\), label='(.*?)', object=Span.*?text='(.*?)'.*?confidence=(\d\.\d+)"
    matches = re.findall(pattern, entry)
    return [
        {
            "Subject": m[0],
            "Predicate": m[1],
            "Object": m[2],
            "Confidence": float(m[3]),
        }
        for m in matches
    ]

# Apply parsing to Relik data
relik_data['relationships_extracted'] = relik_data['relationships'].apply(
    lambda x: parse_relik_relationships(x) if isinstance(x, str) else []
)

# Flatten extracted relationships into a new DataFrame
relik_flat = [
    {
        **item,
        "Source": row[source_column_name],  # Use the first column dynamically as the Source
        "CSV": row["CSV"]                  # Include the CSV file name
    }
    for _, row in relik_data.iterrows()
    for item in row['relationships_extracted']
]

# Create a DataFrame from extracted relationships
relik_df = pd.DataFrame(relik_flat, columns=['Subject', 'Predicate', 'Object', 'Confidence', 'Source', 'CSV'])

# Step 4: Normalize Subject and Object
def normalize_text(text):
    return re.sub(r'\s+', ' ', re.sub(r'[^\w\s]', '', text.lower())).strip()

relik_df['Normalized_Subject'] = relik_df['Subject'].apply(normalize_text)
relik_df['Normalized_Object'] = relik_df['Object'].apply(normalize_text)

# Step 5: Remove Pronouns
pronouns = {
    "i", "me", "my", "mine", "myself",
    "you", "your", "yours", "yourself",
    "he", "him", "his", "himself",
    "she", "her", "hers", "herself",
    "it", "its", "itself",
    "we", "us", "our", "ours", "ourselves",
    "they", "them", "their", "theirs", "themselves",
    "this", "that", "these", "those",
    "who", "whom", "whose", "which", "what",
    "somebody", "someone", "something",
    "anybody", "anyone", "anything",
    "everybody", "everyone", "everything",
    "nobody", "no one", "nothing",
}

def remove_pronouns(text):
    if not isinstance(text, str):
        return text
    words = text.split()
    cleaned_words = [word for word in words if word.lower() not in pronouns]
    return " ".join(cleaned_words).strip()

relik_df["Subject"] = relik_df["Subject"].apply(remove_pronouns)
relik_df["Object"] = relik_df["Object"].apply(remove_pronouns)

# Remove rows where Subject or Object is empty after pronoun removal
relik_df = relik_df[(relik_df["Subject"] != "") & (relik_df["Object"] != "")]

# Step 6: Standardize Nationalities and Countries
def standardize_country(entity):
    if not isinstance(entity, str) or entity.strip() == "":
        return entity  # Skip empty or non-string values

    entity_lower = entity.lower().strip()

    # Check if it's in the demonyms.csv mapping
    if entity_lower in demonyms_mapping:
        return demonyms_mapping[entity_lower]

    # Check if it's a valid country using pycountry
    try:
        country = pycountry.countries.lookup(entity)
        return country.name  # Return the official country name
    except LookupError:
        return entity  # Return original if no match

relik_df["Subject"] = relik_df["Subject"].apply(standardize_country)
relik_df["Object"] = relik_df["Object"].apply(standardize_country)

# Step 7: Remove Duplicate Relationships (Regardless of Predicate)
relik_df = relik_df.drop_duplicates(subset=["Subject", "Object"])

# Step 8: Remove Circular Relationships (A → B and B → A)
def is_circular(row, relationships_set):
    subj, obj = row["Subject"], row["Object"]
    return (obj, subj) in relationships_set

relationships_set = set(zip(relik_df["Subject"], relik_df["Object"]))
relik_df["Is_Circular"] = relik_df.apply(lambda row: is_circular(row, relationships_set), axis=1)
relik_df = relik_df[~relik_df["Is_Circular"]]

# Drop helper columns
relik_df = relik_df.drop(columns=["Is_Circular","Normalized_Subject","Normalized_Object","CSV"])

# Step 9: Export Clean Data
output_file = "relik_cleaned_data.csv"
relik_df.to_csv(output_file, index=False)

print(f"Processed Relik data saved to '{output_file}'")
print(relik_df[['Subject', 'Predicate', 'Object', 'Confidence', 'Source']].head())

# Download the file
files.download(output_file)

```

```python
# Install required libraries
!pip install pandas neo4j

# Import necessary libraries
import pandas as pd
from neo4j import GraphDatabase

# Upload the CSV file
from google.colab import files
uploaded = files.upload()  # Select your file

# Read the CSV into a DataFrame
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)

# The new CSV structure has columns: subject, predicate, object, confidence, source, PDF path
# Extract only the relevant columns for Neo4j: subject, predicate, object
relationships_df = data[['Subject', 'Predicate', 'Object']]

# Display the DataFrame
print("Extracted Relationships:")
print(relationships_df.head())

# Connect to the Neo4j database
NEO4J_URI = #
NEO4J_USERNAME = #
NEO4J_PASSWORD = #

# Create a connection to the Neo4j database
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))

# Function to add nodes and relationships in Neo4j
def add_relationships_to_neo4j(tx, subject, predicate, obj):
    query = f"""
    MERGE (s:Entity {{name: $subject}})
    MERGE (o:Entity {{name: $object}})
    MERGE (s)-[:`{predicate}`]->(o)
    """
    tx.run(query, subject=subject, object=obj)


# Load relationships into Neo4j
with driver.session() as session:
    for _, row in relationships_df.iterrows():
        session.write_transaction(add_relationships_to_neo4j, row['Subject'], row['Predicate'], row['Object'])

print("Relationships successfully added to Neo4j!")

# Close the Neo4j driver connection
driver.close()
```
